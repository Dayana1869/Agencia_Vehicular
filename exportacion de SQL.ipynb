# [dbo].[Vw_AVehiculos]

```sql
/****** Object:  View [dbo].[Vw_AVehiculos]    Script Date: 8/03/2025 01:42:47 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create view [dbo].[Vw_AVehiculos]
	as
	(SELECT c.Nombre,c.Apellido,v.Monto as PRECIO_VEHICULOfrom tblClientes cinner join tblVentas v 	on  c.ClienteID = v.ClienteIDwhere Monto > 20000);
GO
```

# [dbo].[Vw_BVehiculos]

```sql
/****** Object:  View [dbo].[Vw_BVehiculos]    Script Date: 8/03/2025 01:42:47 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create view [dbo].[Vw_BVehiculos]
	as
	(
SELECT 
    ClienteID, 
    nombre, 
    apellido, 
		(SELECT SUM(Monto) 
		FROM tblVentas v 
		WHERE c.ClienteID = v.ClienteID) AS MONTO_GASTADO
FROM tblClientes c
);
GO
```

# [dbo].[Vw_CVehiculos]

```sql
/****** Object:  View [dbo].[Vw_CVehiculos]    Script Date: 8/03/2025 01:42:47 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create view [dbo].[Vw_CVehiculos]
	as
	(
--OPC 1SELECT  C.Nombre, VH.Marca, v.MontoFROM tblClientes CINNER JOIN tblVentas V ON C.ClienteID = V.ClienteIDINNER JOIN tblVehiculos VH ON V.VehiculoID = VH.VehiculoID);
GO
```

# [dbo].[Vw_C2Vehiculos]

```sql
/****** Object:  View [dbo].[Vw_C2Vehiculos]    Script Date: 8/03/2025 01:42:47 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create view [dbo].[Vw_C2Vehiculos]
	as
	(--OPC 2SELECT  V.VentaID, V.VehiculoID, V.ClienteID, C.Nombre, VH.Marca AS Tipo_Vehiculo, VH.Modelo, V.FechaVenta, v.MontoFROM tblClientes CINNER JOIN tblVentas V ON C.ClienteID = V.ClienteIDINNER JOIN tblVehiculos VH ON V.VehiculoID = VH.VehiculoID);
GO
```

# [dbo].[Vw_DVehiculos]

```sql
/****** Object:  View [dbo].[Vw_DVehiculos]    Script Date: 8/03/2025 01:42:47 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create view [dbo].[Vw_DVehiculos]
	as
	(SELECT * 
FROM  tblClientes
WHERE ClienteID IN (
    SELECT ClienteID 
    FROM tblVentas 
    GROUP BY ClienteID 
    HAVING COUNT(*) > 1)
);
GO
```

# [dbo].[usp_vehiculos_editar]

```sql
/****** Object:  StoredProcedure [dbo].[usp_vehiculos_editar]    Script Date: 8/03/2025 01:42:47 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- =============================================
-- Author:		<Author,,Name>
-- Create date: <Create Date,,>
-- Description:	<Description,,>
-- =============================================
CREATE PROCEDURE [dbo].[usp_vehiculos_editar]
	(

	@Codigo int,
	@Marca varchar (50),
	@Modelo varchar(50),
	@Año int,
	@Precio decimal,
	@Estado varchar(15)

	)
AS
BEGIN
	SET NOCOUNT ON;

	 Update tblVehiculos
	set Marca = @Marca,
		Modelo = @Modelo,
		Año = @Año,
		Precio = @Precio,
		Estado = @Estado
	where VehiculoID = @Codigo

END
GO
```
